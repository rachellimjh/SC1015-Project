```python
with open('communities.txt', 'r') as f:
    headers = [line.strip() for line in f.readlines()] # read header names from file into a list

communitiesData = pd.read_csv('communities.csv', header=None) # specify that there are no headers in the original file
communitiesData.columns = headers # set the column headers to the list of headers
communitiesData.to_csv('communities.csv', index=False) # write the DataFrame with headers to the same CSV file

communitiesData.head()
```
